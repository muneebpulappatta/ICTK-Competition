# ENCODING

Nominal data:- Data are not in any order,One Hot Encoding is used in this case


Ordinal data:- Data are in order,Label Encoding is used in this case

Continent is a Nominal values, One Hot Encoding is used.

```python
continent = pd.get_dummies(df["continent"], drop_first= True)
```

```python
continent.tail()
```

Location is a Nominal values, One Hot Encoding is used

```python
location = pd.get_dummies(df["location"], drop_first= True)
```

```python
location.head()
```

ISO_Code is a Nominal values, One Hot Encoding is used

```python
iso_code = pd.get_dummies(df["location"], drop_first= True)
```

```python
iso_code.head()
```

Life expectancy is an ordinal value, hence Label encoding is used.

```python
from sklearn.preprocessing import LabelEncoder
label_en=LabelEncoder()
life_expectancy=label_en.fit_transform(df['life_expectancy'])
life_expectancy=pd.DataFrame(life_expectancy,columns=['life_expectancy'])
life_expectancy
```

Date is an ordinal value, hence Label encoding is used.

```python
from sklearn.preprocessing import LabelEncoder
label_en=LabelEncoder()
date=label_en.fit_transform(df['date'])
date=pd.DataFrame(date,columns=['date'])
date
```

```python
df = pd.concat([df, continent, location,iso_code,life_expectancy,date], axis = 1)
df.head()
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.drop(["continent", "location", "iso_code","life_expectancy","date"], axis = 1, inplace = True)
df.head()
```

# MODELLING

```python
#spilitting data 
y=df["total_cases"]
x=df.drop(["total_cases",],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.25)
df.dtypes
```

# Linear Regression

```python
from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
predictions=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
print("MSE: ",mean_squared_error(y_test,predictions))
```

```python
from sklearn.metrics import r2_score
print("R square value: ",r2_score(y_test,predictions))
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
reg_rf = RandomForestRegressor()
reg_rf.fit(x_train, y_train)
predictions=reg_rf.predict(x_test)
```

```python
print("MSE: ",mean_squared_error(y_test,predictions))
```

```python
print("R square value: ",r2_score(y_test,predictions))
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
dt_model=DecisionTreeRegressor()
dt_model.fit(x_train, y_train)
predictions=dt_model.predict(x_test)
```

```python
print("MSE: ",mean_squared_error(y_test,predictions))
```

```python
print("R square value: ",r2_score(y_test,predictions))
```

# MODEL FINE TUNING

```python
from sklearn.ensemble import RandomForestRegressor
reg_rf = RandomForestRegressor()
reg_rf.fit(x_train, y_train)
predictions=reg_rf.predict(x_test)
```

```python
print("MSE: ",mean_squared_error(y_test,predictions))
print("R square value: ",r2_score(y_test,predictions))
```

From the above models its evident that, Random Forest Regressor with hyper parameter tuning gives the best fit.It has the least MSE value(MSE=211385.72826316796) and maximum R square value (R-square=0.9961660205283639)
